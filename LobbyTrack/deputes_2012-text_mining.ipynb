```python
import base64
import io
import gzip
import re
from bs4 import BeautifulSoup

def strip_html(text):
    if(text is None):
        return "";
    soup = BeautifulSoup(text)
    return soup.get_text()

def add_normal_text(df_out, df_in, idcol , label, text):
    df = pd.DataFrame();
    df['ID'] = df_in[idcol].apply(str);
    df['Type'] = label;
    df['Text'] = df_in[text].apply(strip_html);
    print(df.head())
    df_out = pd.concat([df_out, df] , axis = 0);
    return df_out

def decode_text(text):
    try:
        # print("TEXT_DECODE" , text)    
        # text_data = base64.b64decode(text)
        text_data = base64.urlsafe_b64decode(text)
        buf = io.BytesIO(text_data)
        gzip_f = gzip.GzipFile(fileobj=buf)
        content = gzip_f.read()
        out = content.decode("utf-8")
    except:
        out = "";
    return out;

def add_encoded_text(df_out, df_in, idcol , label, text):
    df = pd.DataFrame();
    df['ID'] = df_in[idcol].apply(str);
    df['Type'] = label;
    lDecoded = df_in[text].apply(decode_text);
    df['Text'] = lDecoded.apply(strip_html);
    print(df.head())
    df_out = pd.concat([df_out, df] , axis = 0);
    return df_out

def build_vocabulary(tables):
    df_out = pd.DataFrame();
    df_out = add_normal_text(df_out, tables['question_ecrite'], 'id' , 'question_ecrite' , 'question') 
    df_out = add_normal_text(df_out, tables['alinea'], 'id' , 'alinea' , 'texte') 
    df_out = add_normal_text(df_out, tables['amendement'], 'id' , 'amendement_texte' , 'texte') 
    df_out = add_normal_text(df_out, tables['amendement'], 'id' , 'amendement_expose' , 'expose') 
    df_out = add_normal_text(df_out, tables['intervention'], 'id' , 'intervention' , 'intervention') 
    df_out = add_normal_text(df_out, tables['titre_loi'], 'id' , 'titre_loi_titre' , 'titre') 
    df_out = add_normal_text(df_out, tables['titre_loi'], 'id' , 'titre_loi_expose' , 'expose') 
    df_out = add_normal_text(df_out, tables['article_loi'], 'id' , 'article_loi_expose' , 'expose') 
    df_out = add_encoded_text(df_out, tables['texteloi'], 'id' , 'texteloi_contenu' , 'contenu') 
    print(df_out.head())
    return df_out;




```
