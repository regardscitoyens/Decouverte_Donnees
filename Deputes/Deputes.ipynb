```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("professions_enriched.csv");
```

```python
df.sample(12)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(df['profession'].values)
X_train_counts.shape
```

```python
count_vect.vocabulary_
```

```python
pd.DataFrame(X_train_counts.toarray()).sample(10)
```

```python
#count_vect.vocabulary_.items()
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tf_transformer = TfidfTransformer(use_idf=False).fit(X_train_counts)
X_train_tf = tf_transformer.transform(X_train_counts)
X_train_tf.shape
```

```python
pd.DataFrame(X_train_tf.toarray())
```

```python
from sklearn import metrics
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
```

```python
targets = ['statut', 'public_prive', 'fonction'];
df_train = df[df['fonction'] != '??']
for tgt in targets:
    print(tgt);
    clf = Pipeline([('vect', CountVectorizer()),
                         ('tfidf', TfidfTransformer()),
                         ('clf', MultinomialNB()),
    ])
    clf = clf.fit(df_train['profession'].values, df_train[tgt].values)
    print(clf.score(df_train['profession'].values, df_train[tgt].values))
    predicted = clf.predict(df['profession'].values)
    df['predicted_' + tgt] = predicted
    print(metrics.confusion_matrix(df[tgt].values, df['predicted_' + tgt].values))
```
